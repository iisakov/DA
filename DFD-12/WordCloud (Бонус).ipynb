# WordCloud

WordCloud (облако слов/тегов) — это визуальное представление списка категорий или тегов, также называемых метками, ярлыками, ключевыми словами и т. п. Обычно используется для описания тегов (метаданных) на веб-сайтах или для представления неформатированного текста. 

Ключевые слова чаще всего представляют собой отдельные слова, важность каждого ключевого слова обозначается размером шрифта или цветом. Такое представление удобно для быстрого восприятия наиболее известных терминов и для распределения терминов по популярности относительно друг друга.

[wikipedia.org](https://ru.wikipedia.org/wiki/Облако_тегов)

### Хохломская роспись

```python
hohloma_mask = Image.open('/content/hohloma_mask.png')
```

```python
plt.imshow(hohloma_mask)
```

```python
hohloma_mask_array = np.array(hohloma_mask)
```

```python
hohloma_df = pd.read_csv('/content/hohloma_words.csv')
hohloma_df = hohloma_df['Названия'].value_counts()
hohloma_df.to_csv('hohloma_df.csv')
```

```python
# Цвет слов основан на маске.
# Цвет фона белый.
wordcloud.WordCloud(background_color = 'white',
                    mode = 'RGB',
                    width = 3000,
                    height = 3000,
                    max_font_size=40,
                    min_font_size=1,
                    max_words=10000,
                    mask=hohloma_mask_array,
                    color_func=wordcloud.ImageColorGenerator(image=hohloma_mask_array)).generate_from_frequencies(hohloma_df).to_file('hohloma_result_background.png')
```

```python
plt.imshow(wordcloud.WordCloud(background_color = 'white',
                    mode = 'RGB',
                    width = 3000,
                    height = 3000,
                    max_font_size=40,
                    min_font_size=1,
                    max_words=10000,
                    mask=hohloma_mask_array,
                    color_func=wordcloud.ImageColorGenerator(image=hohloma_mask_array)).generate_from_frequencies(hohloma_df))
```

```python
# Цвет слов основан на маске.
# Цвет фона прозрачный.
wordcloud.WordCloud(background_color = None,
                    mode = 'RGBA',
                    width = 3000,
                    height = 3000,
                    max_font_size=40,
                    min_font_size=1,
                    max_words=10000,
                    mask=hohloma_mask_array,
                    color_func=wordcloud.ImageColorGenerator(image=hohloma_mask_array)).generate_from_frequencies(hohloma_df).to_file('hohloma_result_non_contour.png')
```

```python
# Цвет слов трёх цветов.
# Цвет фона прозрачный.

# Красный
wordcloud.WordCloud(background_color = None,
                    mode = 'RGBA',
                    width = 3000,
                    height = 3000,
                    max_font_size=40,
                    min_font_size=1,
                    max_words=10000,
                    mask=hohloma_mask_array,
                    color_func=lambda *args, **kwargs: 'red').generate_from_frequencies(hohloma_df).to_file('hohloma_result_non_contour_red_words.png')

# зелёный
wordcloud.WordCloud(background_color = None,
                    mode = 'RGBA',
                    width = 3000,
                    height = 3000,
                    max_font_size=40,
                    min_font_size=1,
                    max_words=10000,
                    mask=hohloma_mask_array,
                    color_func=lambda *args, **kwargs: 'green').generate_from_frequencies(hohloma_df).to_file('hohloma_result_non_contour_green_words.png')

# желтый
wordcloud.WordCloud(background_color = None,
                    mode = 'RGBA',
                    width = 3000,
                    height = 3000,
                    max_font_size=40,
                    min_font_size=1,
                    max_words=10000,
                    mask=hohloma_mask_array,
                    color_func=lambda *args, **kwargs: 'yellow').generate_from_frequencies(hohloma_df).to_file('hohloma_result_non_contour_yellow_words.png')
```
